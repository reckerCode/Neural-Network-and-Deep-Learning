```python
# importing necessary libraries
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
import time 
import math
```

```python
epochs = 100
b = 0
error = 0
```

```python
class lms():
  def __init__(self):
    np.random.seed(69)
    self.weights = 2 * np.random.random((2,1)) - 1
  
  def train_lms(self, inputs, results, epoch):
   lr_rate = 0.001
   for i in range(epoch):
      output = self.preditct(inputs)
      error = results - output
      n_weight = lr_rate * np.dot(inputs.T, error)
      self.weights += n_weight
      print("Training RMSE: ", np.square(error).mean())
      matched = 0
      if results[i] == output[i]:
        matched += 1
      accuracy = matched / float(len(results)) * 100    
      print("Training Accuracy:", accuracy)
      return output
  
  def activation(self, x):
    a = 1/(1+np.exp(-x))
    a[a >= 0.5] = 1
    a[a < 0.5] = 0
    return a
            
  def preditct(self,inputs):
    np.set_printoptions(suppress = True)
    output = self.activation(np.dot(inputs.astype(float), self.weights))
    return output
    
  def accuracy(self,oglLabel,prediction):
    matched = 0
    for i in range(len(oglLabel)):
      if oglLabel[i] == prediction[i]:
        matched += 1
    accuracy = matched / float(len(oglLabel))      
    return accuracy
```

```python
dataset = pd.read_csv("Downloads/transfusion.data") 
```

```python
dataset
```

```python
xx = (pd.DataFrame(dataset, columns = ['Recency (months)', 'Frequency (times)']).to_numpy())
yy = (pd.DataFrame(dataset, columns = ['whether he/she donated blood in March 2007']).to_numpy())
x_train, x_test, y_train, y_test = train_test_split(xx, yy, shuffle = True, random_state = 33)
```

```python
t1 = time.time()
print("Training LMS...")
sp1 = lms()
sp1.train_lms(x_train, y_train, epochs)
print("Training Completed")
# print("Training RMSE:", err)
t2 = time.time()

t3 = time.time()
model_prediction = sp1.preditct(x_test)
error_rate = np.square(np.subtract(y_test,model_prediction)).mean()
accuracy = sp1.accuracy(y_test,model_prediction)
```

```python
print()
print("Testing RMSE :   ",(error_rate))
print("Testing Accuracy :", (accuracy))
t4 = time.time()
print()
print("Training Time: ", (t2-t1))
print("Testing Time :", (t4-t3))
print("Total time :", (t4-t1))
```
